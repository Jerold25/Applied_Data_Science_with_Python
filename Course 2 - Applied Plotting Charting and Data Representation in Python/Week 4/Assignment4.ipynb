Looking for an example? Here's what our course assistant put together for the **Ann Arbor, MI, USA** area using **sports and athletics** as the topic. [Example Solution File](./readonly/Assignment4_example.pdf)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading both the csv files into dataframes
df1 = pd.read_csv('Wage by Gender in Common Jobs - Ann Arbor.csv')
df2 = pd.read_csv('Wage by Gender in Common Jobs - USA.csv')

# Choosing the required columns
df_columns = ['ID Gender', 'Gender', 'Year', 'Detailed Occupation', 'Average Wage', 'Average Wage Appx MOE', 'Total Population','Total Population MOE Appx', 'Geography' ]
df1 = df1[df_columns]
df2 = df2[df_columns]

# Seperating out male and female data in Ann arbor
Ann_Male = df1[df1['Gender'] == 'Male']
Ann_Female = df1[df1['Gender'] == 'Female']

# Seperating out male and female data across all states in USA
Usa_Male = df2[df2['Gender'] == 'Male']
Usa_Female = df2[df2['Gender'] == 'Female']

# Computing the average wage in each occupation of male and female in Ann Arbor from year 2013 to 2017
Ann_Male_mean = Ann_Male.groupby('Detailed Occupation')['Average Wage'].agg({'Avg_mmean': np.mean})
Ann_Female_mean = Ann_Female.groupby('Detailed Occupation')['Average Wage'].agg({'Avg_fmean': np.mean})

# Computing the average wage in each occupation of male and female in all states of USA from year 2013 to 2017
Usa_Male_mean = Usa_Male.groupby('Detailed Occupation')['Average Wage'].agg({'Avg_mmean': np.mean})
Usa_Female_mean = Usa_Female.groupby('Detailed Occupation')['Average Wage'].agg({'Avg_fmean': np.mean})

# Sorting the Average wage in descending order - Ann Arbor 
Ann_Male_mean = Ann_Male_mean.sort_values(by='Avg_mmean', ascending=False).reset_index()
Ann_Female_mean = Ann_Female_mean.sort_values(by='Avg_fmean',ascending=False).reset_index()

# Sorting the Average wage in descending order - USA
Usa_Male_mean = Usa_Male_mean.sort_values(by='Avg_mmean', ascending=False).reset_index()
Usa_Female_mean = Usa_Female_mean.sort_values(by='Avg_fmean',ascending=False).reset_index()

# Plotting a figure
plt.figure()

# Super title of the Plot
plt.suptitle('Comparison of Wages by Gender between Ann Arbor, MI and all states across USA from the year 2014 to 2017.', family = 'serif', fontsize='xx-large', fontweight = 'semibold')

# Configuriing the height and width of the figure
fig_size = plt.rcParams["figure.figsize"] 

# Set figure width to 20 and height to 12
fig_size[0] = 20
fig_size[1] = 12
plt.rcParams["figure.figsize"] = fig_size

# Plotting of 1st bar graph in axis 1
ax = plt.subplot(2,1,1)
plt1 = plt.bar(range(Ann_Male_mean.shape[0]), Ann_Male_mean['Avg_mmean'],width = 0.3, color='green', label = 'Male')

'''
This is for Lollipop graph
(markerline, stemlines, baseline) = plt.stem(range(Ann_Male_mean.shape[0]), Ann_Male_mean['Avg_mmean'],width = 0.3, color='green', label = 'Male')
plt.setp(baseline, visible=False)
plt.setp(stemlines, color='red', linewidth = 1.0)
'''

# plot another set of bars, adjusting the new xvals to make up for the first set of bars plotted - Grouped bar charts(Ann arbor)
new_xvals = []
for item in range(Ann_Female_mean.shape[0]):
    new_xvals.append(item+0.33)

plt2 = plt.bar(new_xvals, Ann_Female_mean['Avg_fmean'],width = 0.3, color='red', label = 'Female')

'''
This is for Lollipop graph
(markerline, stemlines, baseline) = plt.stem(new_xvals, Ann_Female_mean['Avg_fmean'],width = 0.3, color='red', label = 'Female')
plt.setp(baseline, visible=False)
plt.setp(stemlines, color='red', linewidth = 1.0)
'''
# Remove the Right and Top axis lines from graph 1
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# X, Y labels and Titles of the graph 1
plt.xlabel('Occupation', family = 'serif', fontsize='x-large', fontweight = 'semibold', labelpad = 15)
plt.ylabel('Average Full-Time Salary', family = 'serif', fontsize='x-large', fontweight = 'semibold')
ax.set_title('Wages by Gender only in Ann arbor, Michigan, USA', family = 'serif', fontsize='x-large', fontweight = 'semibold', y = 0.93)

# labels for x ticks
reorder = [i + 0.16 for i in range(Ann_Male_mean.shape[0])]
plt.xticks(reorder, Ann_Male_mean['Detailed Occupation'] )

# Legends for graph 1
plt.legend(frameon=False, prop={'size':17})

# Adjusting the height of the plots - such that it's position wont intersect
plt.subplots_adjust(bottom=-0.1)

# Plotting of 1st bar graph in axis 1
bx = plt.subplot(2,1,2)
plt3 = plt.bar(range(Usa_Male_mean.shape[0]), Usa_Male_mean['Avg_mmean'],width = 0.3, color='green', label = 'Male')

'''This is for Lollipop graph
(markerline, stemlines, baseline) = plt.stem(range(Usa_Male_mean.shape[0]), Usa_Male_mean['Avg_mmean'], width = 0.3, color='green', label = 'Male')
plt.setp(baseline, visible=False)
plt.setp(stemlines, color='red', linewidth = 1.0)'''

# plot another set of bars, adjusting the new xvals to make up for the first set of bars plotted - Grouped bar charts(USA)
new_xvals = []
for item in range(Usa_Female_mean.shape[0]):
    new_xvals.append(item+0.33)

plt4 =plt.bar(new_xvals, Usa_Female_mean['Avg_fmean'],width = 0.3, color='red', label = 'Female')

'''
This is for Lollipop graph
(markerline, stemlines, baseline) = plt.stem(new_xvals, Usa_Female_mean['Avg_fmean'], width = 0.3, color='red', label = 'Female')
plt.setp(baseline, visible=False)
plt.setp(stemlines, color='red', linewidth = 1.0)
'''
# Remove the Right and Top axis lines from graph 2
bx.spines['right'].set_visible(False)
bx.spines['top'].set_visible(False)

# X, Y labels and Titles of the graph 2
plt.xlabel('Occupation', family = 'serif', fontsize='x-large', fontweight = 'semibold', labelpad = 20)
plt.ylabel('Average Full-Time Salary', family = 'serif', fontsize='large', fontweight = 'semibold')
bx.set_title('Wages by Gender across all states in USA', family = 'serif', fontsize='x-large', fontweight = 'semibold', y = 0.93)

# labels for x ticks
reorder = [i + 0.16 for i in range(Usa_Male_mean.shape[0])]
plt.xticks(reorder, Usa_Male_mean['Detailed Occupation'] )

# Legends for graph 2
plt.legend(frameon=False, prop={'size':17})


#plt.savefig('plot.png', bbox_inches='tight')
plt.show()

```
